### sqlite3 Demo

#### Definitions

```python
import sqlite3

table_name =	'Students'

fields = 'name TEXT, student_id TEXT, gpa REAL'
field_names = 'name, student_id, gpa'

students = (('Bob', 		'G123456789',	'3.02'),
			('Alice',	'G234567890',	'2.93'),
			('Perry',	'G345678901',	'3.95'),
			('Charlie',	'G456789012',	'1.25'))
```

#### Connect To Database

```python
conn = sqlite3.connect(':memory:')	
crsr = conn.cursor()
```

#### Create SQL Table

```python
sql_cmd = f'CREATE TABLE {table_name} ({fields})'
crsr.execute(sql_cmd)
```

#### Insert Data

```python
for student in students:
	sql_cmd = f'INSERT INTO {table_name} ' \
				f'({field_names}) ' \
				f"VALUES ('{student[0]}', '{student[1]}', '{student[2]}')"
	crsr.execute(sql_cmd)
```

#### Select All Students

```python
sql_cmd = f'SELECT * FROM {table_name}'
crsr.execute(sql_cmd)
selected_students = crsr.fetchall()
for student in selected_students:
    print(student)
```

#### Select Students With GPA >= 3.0

```python
sql_cmd = f'SELECT * FROM {table_name} WHERE gpa >= 3'
crsr.execute(sql_cmd)
selected_students = crsr.fetchall()
for student in selected_students:
    print(student)
```
